Langgraph is a tool that helps to build reliable agents. The more autonomous agents are, the less reliable they are. To bend the graph a little bit, we use langgraph. 

The main difference between langchain and langgraph is that langchain is used to design a linear workflow with little to no autonomy. Whereas langgraph allows autonomous decisions by the LLM, allowing it to decide next steps and what steps to take while making decisions
